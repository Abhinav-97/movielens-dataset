```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
user_columns = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']
```

```python
user_data = pd.read_csv('../data/ml-1m/ml-1m/users.dat', names=user_columns, sep='::')
```

```python
user_data
```

```python
ratings_column = ['UserID', 'MovieID', 'Rating', 'Timestamp']
movies_column = ['MovieID', 'Title', 'Genres']
```

```python
ratings_data = pd.read_csv('../data/ml-1m/ml-1m/ratings.dat', sep='::', names=ratings_column)
```

```python
ratings_data
```

```python
movie_data = pd.read_csv('../data/ml-1m/ml-1m/movies.dat', sep='::', names=movies_column)
```

```python
movie_data
```

```python
movies_ratings = pd.merge(movie_data, ratings_data)
user_ratings = pd.merge(movies_ratings, user_data)
```

```python
user_ratings
```

```python
most_rated = user_ratings.groupby('Title')
```

```python
mean_ratings = user_ratings.groupby('Title').agg({'Rating':[np.size, np.mean]})
```

```python
frequently_reviewed = mean_ratings['Rating']['size'] >=50
```

```python
mean_ratings[frequently_reviewed].sort_values([('Rating', 'mean')], ascending=False)
```

Highest Rated 25 movies

```python
mean_ratings[frequently_reviewed].sort_values([('Rating', 'mean')], ascending=False)[:25]
```

average ratings for different genres according to gender

```python
genre_rating = pd.pivot_table(user_ratings, index=['Genres'], values=['Rating'], columns=['Gender'])
```

get differences between both male and female ratings

```python
genre_rating['diff'] = genre_rating['Rating']['M'] - genre_rating['Rating']['F']
```

```python
genre_rating['new_col']=range(0,301)
```

```python
genre_rating['diff']
```

```python
# genre_rating = genre_rating['diff'].astype(float)
genre_rating.diff.values
```

```python
most_negetive = genre_rating.sort_values('diff')[:25]
most_positive = genre_rating.sort_values('diff', ascending=False)[:25]
difference = pd.concat([most_positive, most_negetive]) 
```

```python
difference = difference['diff']
difference
```
