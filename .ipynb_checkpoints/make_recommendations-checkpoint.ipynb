```python
ratings_column = ['UserID', 'MovieID', 'Rating', 'Timestamp']
movies_column = ['MovieID', 'Title', 'Genres']
```

```python
movie_data = pd.read_csv('../data/ml-1m/ml-1m/movies.dat', sep='::', names=movies_column)
```

```python
movies_ratings = pd.merge(movie_data, ratings_data)
user_ratings = pd.merge(movies_ratings, user_data)
```

```python
user_ratings
```

get rating of each movie from every user (0 is for not rated)

```python
each_movie_rating = pd.pivot_table(user_ratings, index=['UserID'], values=['Rating'], columns=['Title'], fill_value=0)
```

```python
each_movie_rating
```

# Get correlation scores of movies

```python
movie_corr = np.corrcoef(each_movie_rating.T)
```

```python
movie_corr.shape
```

```python
movies_list = list(each_movie_rating)
movies_index = []
```

```python
for i in range(len(movies_list)):
    movies_index.append(movies_list[i][1])
```

```python
movies_index[::]
```

```python
movies_index.index('101 Dalmatians (1996)')
```

```python
favourite_movie = 'Exorcist, The (1973)'
favourite_movie_index = movies_index.index(favourite_movie)
corr_score  = movie_corr[favourite_movie_index]
```

```python
sorted(corr_score,reverse=True)
```

select only those movies which have high correltaion score


```python
condition = (corr_score >= 0.4)
```

```python
np.extract(condition, movies_index)
```

get correlation score based on the movie

```python
def get_similar_movies(movie_title):
    title_index = movies_index.index(movie_title)
#     print(title_index)
    return movie_corr[title_index]
```

```python
def get_movie_recommendations(user_movies):
    
    movie_similarities = np.zeros(movie_corr.shape[0])
    for movie in user_movies:
#         print(movie)
        similarity_score = get_similar_movies(movie)
#         print(similarity_score)
        movie_similarities += similarity_score
#     print(movie_similarities)
    recommendation = []  
    for i in range(len(movies_index)):
        recommendation.append((movies_index[i],movie_similarities[i]))
        
    return sorted(recommendation, key= lambda x:x[1],reverse=True)
```

```python
user_movies = ['Rocky (1976)', 'Face/Off (1997)', 'Gladiator (2000)', 'Lake Placid (1999)', 'Nightmare on Elm Street, A (1984)', 'Batman Forever (1995)', 'Exorcist, The (1973)']
```

```python
recommendations = get_movie_recommendations(user_movies)
```

```python
recommendations
```

```python
movies_you_would_like = []
for movie,corr_rating in recommendations:
    if movie in user_movies:
        continue
    else:
        movies_you_would_like.append(movie)
        if len(movies_you_would_like) > 10:
            break
```

```python
print("Movies you would like\n")
for m in movies_you_would_like:
    print(m)
```
