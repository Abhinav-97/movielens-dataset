```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
user_columns = ['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code']
```

```python
user_data = pd.read_csv('../data/ml-1m/ml-1m/users.dat', names=user_columns, sep='::')
```

```python
user_data
```

```python
ratings_column = ['UserID', 'MovieID', 'Rating', 'Timestamp']
movies_column = ['MovieID', 'Title', 'Genres']
```

```python
ratings_data = pd.read_csv('../data/ml-1m/ml-1m/ratings.dat', sep='::', names=ratings_column)
```

```python
ratings_data
```

```python
movie_data = pd.read_csv('../data/ml-1m/ml-1m/movies.dat', sep='::', names=movies_column)
```

```python
movie_data
```

```python
movies_ratings = pd.merge(movie_data, ratings_data)
user_ratings = pd.merge(movies_ratings, user_data)
```

```python
user_ratings
```

get average rating for each movie


```python
pd.pivot_table(user_ratings, index=['Title'], values=['Rating'])
```

get average rating for each user

```python
pd.pivot_table(user_ratings, index=['UserID'], values=['Rating'])
```

```python
pd.pivot_table(user_ratings, index=['Age'], values=['Rating'], aggfunc='std')
```

```python
pd.pivot_table(user_ratings, index=['Gender'], values=['Rating'],columns=['Genres'],aggfunc=[np.mean, np.std])
```
